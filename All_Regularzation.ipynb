```python
import numpy as NP
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
from sklearn.linear_model import Ridge
```

```python
df=pd.read_csv('50_Startups.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
X=df.drop(['State','Profit'],axis=1)
Y=df['Profit']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,Y,random_state=0,test_size=0.2)
```

```python
X_train.shape ,X_test.shape
```

```python
lr=LinearRegression()
rd=Ridge(alpha=0.01,solver='sparse_cg')
ls=Lasso(alpha=0.01)
el=ElasticNet(0.0001)
```

```python
lr.fit(X_train,y_train) #Linear Regression
rd.fit(X_train,y_train) #ridge regression
ls.fit(X_train,y_train)  #lasso regression
el.fit(X_train,y_train)  #elastinet regression
```

```python
from sklearn.metrics import r2_score
```

```python
y_pred_lr=lr.predict(X_test)
y_pred_rd=rd.predict(X_test)
y_pred_ls=ls.predict(X_test)
y_pred_el=el.predict(X_test)

r1=r2_score(y_test,y_pred_lr)
r2=r2_score(y_test,y_pred_rd)
r3=r2_score(y_test,y_pred_ls)
r4=r2_score(y_test,y_pred_el)

print(r1,r2,r3,r4)
```
