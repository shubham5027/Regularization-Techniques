```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.metrics import r2_score
```

```python
df=pd.read_csv('titanic_toy.csv')
```

```python
from sklearn.impute import SimpleImputer
Si=SimpleImputer(strategy='mean')
```

```python
df.info()
```

```python
df.describe()
```

```python
X=df.drop(['Survived'],axis=1)
```

```python
Y=df['Survived']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,Y,random_state=0,test_size=0.2)
```

```python
X_train_new=Si.fit_transform(X_train)
X_test_new=Si.transform(X_test)
```

```python
lr=LinearRegression()
rd=Ridge(solver='sparse_cg',alpha=0.1)
ls=Lasso()
```

```python
rd.fit(X_train_new,y_train)  #ridge regressiin9
```

```python
lr.fit(X_train_new,y_train) #Linear Regression
```

```python
ls.fit(X_train_new,y_train) #Lasso regession
```

```python
y_pred_1=lr.predict(X_test_new)
y_pred_2=rd.predict(X_test_new)
y_pred_3=ls.predict(X_test_new)
```

```python
r1=r2_score(y_test,y_pred_1)
r2=r2_score(y_test,y_pred_2)
r3=r2_score(y_test,y_pred_3)

print("linear Regression",r1)
print("ridge regression",r2)
print("Lasss regression",r3)
```
